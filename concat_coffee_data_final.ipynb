# 導入資料、資料合併

```python
import pandas as pd 
import numpy as np
old_data=pd.read_csv('data.csv')
new_data=pd.read_csv('all_coffee_data.csv')

final_data=pd.concat([old_data,new_data],join='inner')
final_data.reset_index(inplace=True, drop=True)
final_data
```

# 輸出CSV檔

```python
#輸出資料
final_data.to_csv('final_coffee_data.csv',index=False)
print(final_data.shape)
```
